```python
# Import Your Libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Part 1

### Your Histogram Equalization Function

```python
def hist_equ(image):
    '''
    input:
    image (ndarray): input image
    output:
    output_image (ndarray): enhanced image
    '''
    
    ############
    # Your code
    # Start
    nk_dict = {}
    nk_sum_dict = {}
    nk_sum_dict[-1] = 0
    max_value = 256
    for i in range(max_value):
        nk_dict[i] = np.count_nonzero(image == i)
        nk_sum_dict[i] = nk_sum_dict[i-1] + nk_dict[i]

    for i in range(1, max_value):       
        nk_sum_dict[i] /= image.size
        nk_sum_dict[i] *=  (max_value - 1)
        nk_sum_dict[i] = round(nk_sum_dict[i])

    output_image = np.zeros_like(image)

    for i in range(max_value):
        output_image = np.where(image != i, output_image, nk_sum_dict[i])


    # End
    
    return output_image
```

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
equ = hist_equ(img)
# END

res = np.hstack((img, equ)) #stacking images side-by-side

plt.figure(figsize=(16, 16))
plt.imshow(res, cmap='gray')
```

## Part 2

### CLAHE OpenCV Library

```python
def CLAHE(image, gridSize, clip_limit):
    '''
    inputs:
    image (ndarray): input image
    gridSize (tuple): window size
    clip_limit (int): threshold for contrast limiting
    output:
    output_image (ndarray): improved image
    '''

    ############
    # Your code
    # Start
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=gridSize)
    output_image = clahe.apply(image)
    # End

    return output_image
```

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
clh = CLAHE(img, clip_limit=2.0, gridSize=(8, 8))
# END

res = np.hstack((img, clh)) #stacking images side-by-side

plt.figure(figsize=(16, 16))
plt.imshow(res, cmap='gray')
```
